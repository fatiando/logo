# Create the Fatiando Data organization logo

Load up a base png with the fatiando logo in black and white. Then convert it to as ASCII art image.

```python
import skimage.io
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
img = skimage.io.imread("fatiando-data-logo-base.png", as_gray=True)
img[img > 0.9] = np.nan
```

```python
plt.imshow(img)
```

```python
def gray2ascii(value):
    if np.isnan(value):
        return " "
    elif value > 0.4:
        return random.choice(["0", "1"])
    else:
        return " "
```

```python
fsize = 8
plt.figure(figsize=(fsize, fsize), facecolor="#08081c")
plt.axis("off")
plt.gca().set_aspect("equal")
for i in range(img.shape[0]):
    for j in range(img.shape[1]):        
        plt.text(j, i, gray2ascii(img[i, j]), fontweight="bold", fontfamily="monospace", color="#7ee03b")
plt.xlim(0, img.shape[1])
plt.ylim(img.shape[0], 0)
plt.savefig("../fatiando-data-logo.png", bbox_inches="tight")
plt.savefig("../fatiando-data-logo.svg", bbox_inches="tight")
```
